```python
import os
import pandas as pd
import torch
from sklearn.model_selection import train_test_split
from torchvision import transforms
from torch.utils.data import DataLoader, Dataset, Subset

from PIL import Image

```

```python
# 自定义 Dataset 类
class CustomDataset(Dataset):
    def __init__(self, csv_file=None, img_dir=None, transform=None, is_test=False):
        if not is_test:

          self.data_frame = pd.read_csv(csv_file)
          self.img_labels = self.data_frame.iloc[:, 1]
          self.classes=sorted(set(self.img_labels))
          self.class_to_idx = {c: i for i, c in enumerate(self.classes)}
             # 训练数据标签
        else:
            self.data_frame = None
            self.img_labels = None  # 测试数据没有标签
        self.img_dir = img_dir
        self.transform = transform
        self.is_test = is_test
        self.img_list = os.listdir(img_dir)




    def __len__(self):
        return len(self.img_list) if self.is_test else len(self.data_frame)

    def __getitem__(self, idx):
        if self.is_test:
            img_name = os.path.join(self.img_dir, self.img_list[idx])
            image = Image.open(img_name)
            if self.transform:
                image = self.transform(image)
            return image, self.img_list[idx]  # 返回图片和文件名用于生成 submission
        else:
            img_name = os.path.join(self.img_dir, self.data_frame.iloc[idx, 0]+'.jpg')
            image = Image.open(img_name)
            label = self.img_labels[idx]
            if isinstance(self.img_labels[idx], str):
              label = self.class_to_idx[self.img_labels[idx]]
            else:
              label = self.img_labels[idx]  # If label is already numerical, no need to convert

            label = torch.tensor(label)
              # 训练数据的标签
            if self.transform:
                image = self.transform(image)
            return image, label

```

```python
# 数据预处理
transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.CenterCrop((224, 224)),# 调整图像大小
    transforms.RandomHorizontalFlip(p=0.5),  # 以50%的概率进行水平翻转
    transforms.RandomRotation(degrees=15),  # 随机旋转15度以内
    #transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # 颜色抖动
    transforms.ToTensor(),          # 转为张量
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # 正规化
])
```

```python
val_transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.CenterCrop((224, 224)),# 调整图像大小
    #transforms.RandomHorizontalFlip(p=0.5),  # 以50%的概率进行水平翻转
    #transforms.RandomRotation(degrees=15),  # 随机旋转15度以内
    #transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # 颜色抖动
    transforms.ToTensor(),          # 转为张量
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # 正规化
])
```

```python
test_transform = transforms.Compose([
    transforms.Resize((256,256)),
    transforms.CenterCrop((224, 224)),# 调整图像大小
    #transforms.RandomHorizontalFlip(p=0.5),  # 以50%的概率进行水平翻转
    #transforms.RandomRotation(degrees=15),  # 随机旋转15度以内
    #transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),  # 颜色抖动
    transforms.ToTensor(),          # 转为张量
    transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])  # 正规化
])
```

```python
# 路径设置
train_csv = 'labels.csv'    # 训练集CSV文件
train_img_dir = 'train/train/'   # 训练集图像文件夹
test_img_dir = 'test/test/'     # 测试集图像文件夹

# 读取数据集
train_dataset = CustomDataset(train_csv, train_img_dir, transform=transform)

# 数据集划分
train_indices, val_indices = train_test_split(
    list(range(len(train_dataset))), test_size=0.15, random_state=42, stratify=train_dataset.data_frame.iloc[:, 1]
)
```

```python
# 创建子集
train_subset = Subset(train_dataset, train_indices)
val_subset = Subset(train_dataset, val_indices)
```

```python
# 加载器
train_loader = DataLoader(train_subset, batch_size=32, shuffle=True, num_workers=2)
val_loader = DataLoader(val_subset, batch_size=32, shuffle=False, num_workers=2)
```

```python
# 读取测试集
test_dataset = CustomDataset(img_dir=test_img_dir, transform=test_transform, is_test=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, num_workers=2)
```

```python
# 检查数据加载器的正确性
print(f"训练集大小: {len(train_subset)}")
print(f"验证集大小: {len(val_subset)}")
print(f"测试集大小: {len(test_dataset)}")
```

```python
#可视化训练集里的图片
# 获取训练集前九个图片
import matplotlib.pyplot as plt
data_iter = iter(train_loader)
images, labels = next(data_iter)

# 显示图像
def imshow(image, ax, title):
    # 将图像从 Tensor 转为 numpy 数组并反归一化
    image = image.numpy().transpose((1, 2, 0))
    mean = [0.485, 0.456, 0.406]
    std = [0.229, 0.224, 0.225]
    image = std * image + mean  # 反归一化
    image = image.clip(0, 1)    # 限制到 [0,1] 之间
    ax.imshow(image)
    ax.set_title(title)
    ax.axis('off')

fig, axes = plt.subplots(3, 3, figsize=(12, 12))
for i, ax in enumerate(axes.flat):
    imshow(images[i], ax, title=f"type: {labels[i]}")

plt.tight_layout()
plt.show()
```

```python

# 使用 EfficientNet 预训练模型
#model = models.efficientnet_b0(pretrained=True)  # EfficientNet-B0 是最小的 EfficientNet 模型
#model = models.efficientnet_b4(pretrained=True)  # EfficientNet-B7 是最大的 EfficientNet 模型

# model = models.resnet1(pretrained=True)


# # 冻结预训练模型的参数
# for param in model.parameters():
#     param.requires_grad = False

# num_frozen_layers = 20  # 假设你想冻结前10个层
# for i, child in enumerate(model.children()):
#     if i < num_frozen_layers:
#         for param in child.parameters():
#             param.requires_grad = False
#     else:
#         for param in child.parameters():
#             param.requires_grad = True

# print(f"前 {num_frozen_layers} 层被冻结，后面的层解冻。")



# # 获取分类层的输入特征数
# num_features = model.fc.in_features  # 获取分类层的输入特征数

# # 更改最后的分类层来适应120个品种
# model.fc = nn.Linear(num_features, 120)  # 假设有120个品种
import torch
import torch.nn as nn
from torchvision import models
from torchvision.models import ResNet101_Weights

import torch
import torch.nn as nn
from torchvision import models

class CustomResNetModel(nn.Module):
    def __init__(self, num_classes=120):
        super(CustomResNetModel, self).__init__()

        # 加载预训练的 ResNet-152 模型
        self.resnet = models.resnet101(ResNet101_Weights.DEFAULT)

        for para in self.resnet.parameters():
          para.requires_grad=False

        num_layer=8
        for i,child in enumerate(self.resnet.children()):
          if i<num_layer:
            for param in child.parameters():
              param.requires_grad=False




        # 去掉 ResNet-152 的最后一个全连接层
        num_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Identity()  # 使用 nn.Identity() 替换掉最后的全连接层

        # 自定义分类器
        self.classifier = nn.Sequential(
            nn.Linear(num_features, 512),  # 输入维度为 ResNet-152 的输出特征数
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)  # 输出维度为类别数（如 120 个狗的种类）
        )

    def forward(self, x):
        # 使用 ResNet-152 的特征提取层
        x = self.resnet(x)
        # 使用自定义的分类层
        x = self.classifier(x)
        return x





model=CustomResNetModel()
# 将模型转移到 GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)


for i, layer in enumerate(list(model.resnet.children())):
  print(f"Layer {i}: {layer}")
```

```python
# 定义损失函数和优化器
import torch.optim as optim
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 训练模型
def train(model, train_loader, val_loader, criterion, optimizer, num_epochs=10):
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for images, labels in train_loader:
            images, labels = images.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

        print(f"Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}")

        # 验证模型
        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(device), labels.to(device)
                outputs = model(images)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()

        print(f"Validation Accuracy: {100 * correct / total}%")

    print("训练完成！")

# 开始训练
train(model, train_loader, val_loader, criterion, optimizer, num_epochs=15)
```

```python
import torch
import pandas as pd

# 假设你的训练数据集为 train_dataset
# 获取训练数据集中的类别名
class_names = train_dataset.classes  # 获取类别名称列表
num_classes = len(class_names)  # 确定类别数量

model.eval()  # 设置模型为评估模式

# 初始化存储图像 ID 和预测结果的列表
image_ids = []
predictions = []

# 遍历测试数据加载器并预测每个图像的类别概率
with torch.no_grad():
    for images, img_names in test_loader:
        images = images.to(device)
        outputs = model(images)  # 获取模型输出

        # 将模型的输出转换为概率
        probabilities = torch.nn.functional.softmax(outputs, dim=1)

        # 将概率转换为 NumPy 数组
        probabilities = probabilities.cpu().numpy()

        # 去除扩展名 '.jpg' 并添加到图像 ID 列表中
        img_names = [name.replace('.jpg', '') for name in img_names]
        image_ids.extend(img_names)

        # 将每个图像的概率添加到预测列表中
        predictions.extend(probabilities)

# 创建 DataFrame
columns = ['id'] + class_names  # 第一列为图像 ID，其他列为类别名称
data = [[image_ids[i]] + list(predictions[i]) for i in range(len(image_ids))]
submission_df = pd.DataFrame(data, columns=columns)

# 对 id 列按字母顺序升序排序
submission_df = submission_df.sort_values(by='id')

# 保存为 CSV 文件
submission_df.to_csv('submission.csv', index=False)

print("提交文件 submission.csv 已生成！")
```
